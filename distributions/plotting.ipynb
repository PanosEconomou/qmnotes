# Plotting QM1 Midterm 1 Grades

```python
#import librries

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
# Get the grades
file    = open("./midterm-grades.csv")
grades  = np.array([float(line) for line in file.readlines()])
file.close()
```

```python
# Some manipulation
lims        = (70,150)
N           = 1000
grade_range = np.linspace(*lims,N)
kernel      = gaussian_kde(grades,bw_method=.5)
density     = kernel(grade_range)
mean        = np.mean(grades)
```

```python
# Plotting time
fig = plt.figure(figsize=(16,8),dpi =200)
ax  = fig.add_subplot(111)
ax.hist(grades,ls='--',lw=1,histtype='step',edgecolor='grey',density=True,label='Histogram')
ax.plot(grade_range,density,c='darkblue', label = 'KDE Density')
ax.plot(grades, np.full_like(grades,-0.001), '|k', markeredgewidth=1,label='Grades')
ax.axvspan(lims[0],105,edgecolor=None,alpha=0.05,color='deepskyblue',lw=0,hatch='///',label=r'$\leq 100\%$')
ax.axvline(mean,c='indigo',ls='--',label=r'Mean $= 120\%$')
ax.legend(frameon=False)
ax.plot()

# Prettification
ax.set_xlabel('Grade')
ax.set_ylabel('Density')
ax.set_yticks([])
ax.spines[['top', 'right']].set_visible(False)
ax.set_xlim(*lims)
```
